```python
import cv2 as cv
from skimage import io
from google.colab.patches import cv2_imshow # for image display
import matplotlib.pyplot as plt
```

```python
img_urls = ["https://media.istockphoto.com/photos/coins-collection-isolated-on-white-picture-id147063335?k=20&m=147063335&s=612x612&w=0&h=KLZ8GF9AjbjoK4OXTcZjnZDogT-hV6Hr7atjczfFhTI="]
```

```python
image_bgr = io.imread(img_urls[0])
image_rgb = cv.cvtColor(image_bgr, cv.COLOR_BGR2RGB)
cv2_imshow(image_rgb)
```

```python
image_gray = cv.cvtColor(image_rgb, cv.COLOR_RGB2GRAY)
cv2_imshow(image_gray)
```

Histogram


```python
#histogram of the rgb image
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([image_rgb],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()

#histogram of the grayscale image
plt.hist(image_gray.ravel(),bins = 256, range = [0,256]) 
plt.show()
```

```python
# Find a good threshold value
ret, thresh = cv.threshold(image_gray, 0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
print('Value of Threshold:', ret)

# Show the binary image 
plt.imshow(thresh, cmap='gray')
plt.show()
```
